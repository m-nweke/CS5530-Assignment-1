```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/School/Raw_yield_data.csv')
df
```

```python
# Use one hot encoding to convert the "Frailty" column
df['Frailty'] = df['Frailty'].replace({'N': 0, 'Y': 1})

# Save the cleaned data to a new CSV file
df.to_csv('/content/drive/MyDrive/School/Cleaned_yield_data.csv', index=False)
df
```

```python
# Calculate the correlation between height and grip strength
heightCorr = df[['Height (Inches)', 'Grip Strength']].corr()

print(heightCorr)
```

```python
# Calculate the correlation between height and grip strength
weightCorr = df[['Weight (pounds)', 'Grip Strength']].corr()

print(weightCorr)
```

```python
# Drop the 'Height (Inches)' and 'Weight (Pounds)' columns
df = df.drop(['Height (Inches)', 'Weight (pounds)'], axis=1)

# Save the result to a new CSV file
df.to_csv('/content/drive/MyDrive/School/cleaned_data.csv', index=False)
df
```
